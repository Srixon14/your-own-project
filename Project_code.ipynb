# ANALYSIS

Finally, this graph represents the evolution of the population of threatened species and the World human population (divided by 1,000,000) over the past 20 years.

As we can see, apparently both variables are pretty related. Seems to be a big correlation between them. And is actually like that: the correlation coefficient is 0.976, which means there's probably a huge cause-effect relation as human population increase means demographic expansion, and therefore habitat destruction for many animal species.

```python
import geopandas as gpd
import os
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.set_style("dark", {"axes.facecolor": ".9"})
shapefile = os.path.expanduser('ne_10m_admin_0_countries/ne_10m_admin_0_countries.shp')

# Values will be binned and separated in as many colors as the scalar we assign this variable.
colors = 20

# cmap refers to the range of colors our bins will take.
cmap = 'OrRd'
figsize = (30, 10)
year = '2005'
title = 'Suicide rates by country in {}'.format(year)


gdf = gpd.read_file(shapefile)[['ADM0_A3', 'geometry']].to_crs('+proj=robin')
gdf.sample(5)

merged = gdf.merge(new_table, left_on='ADM0_A3', right_on='Code')
merged.head()

ax = merged.dropna().plot(column='Total*', cmap=cmap, figsize=figsize, scheme='equal_interval', k=colors, legend=True)
plt.show()

merged[merged.isna().any(axis=1)].plot(ax=ax, color='#fafafa', hatch='///')


ax.set_axis_off()
ax.set_xlim([-1.5e7, 1.7e7])
ax.get_legend().set_bbox_to_anchor((.12, .4))
sns.despine(top=True, bottom=True, right=True, left=True)
```
